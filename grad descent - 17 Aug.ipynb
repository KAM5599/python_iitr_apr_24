```python
import random
import matplotlib.pyplot as plt
```

```python
X = []
Y = []
for x in range(1, 10):
    y = 10*x + 10 + 5*random.random() - 5*random.random()
    X.append(x)
    Y.append(y)
```

```python
plt.scatter(X, Y)
```

```python
def sqloss(weights):
    m, c = weights
    totalsqerror = 0
    for i, x in enumerate(X):
        predictedy = x*m + c
        actualy = Y[i]
        error = actualy - predictedy
        totalsqerror += error * error
    return totalsqerror
```

```python
def sq(x):
    return x*x
```

```python
def diff(f, x):
    delta = 0.000000001
    y = f(x)
    y1 = f(x+delta)
    return (y1 - y)/delta
```

```python
diff(sq, 10)
```

```python
def cube(x):
    return x*x*x
```

```python
diff(cube, 12)
```

```python
def diff(lossf, weights, idx):
    weights = weights.copy()
    delta = 0.000000001
    y = lossf(weights)
    weights[idx] += delta 
    y1 = lossf(weights)
    return (y1 - y)/delta
```

```python
diff(sqloss, [10, 10], 0)
```

```python
diff(sqloss, [10, 10], 1)
```

```python
diff(sqloss, [1, 10], 0)
```

```python
m = random.random()
c = random.random()
```

```python
eta = .002 #Learning rate scheduling
```

```python
for epoch in range(1000):
    weights = [m, c]
    print(epoch, "Current Loss: ", sqloss(weights), weights)
    de_dm = diff(sqloss, weights, 0)
    de_dc = diff(sqloss, weights, 1)
    m1 = m - eta * de_dm
    c1 = c - eta * de_dc
    m, c = m1, c1
```
